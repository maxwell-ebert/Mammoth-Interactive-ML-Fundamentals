#**Calculating Error of Machine Learning Model**

```python
import numpy

yhat_minus_y = numpy.linspace(-5,5, 100)

absolute_error = numpy.abs(yhat_minus_y)

mean_squared_error = yhat_minus_y ** 2

import matplotlib.pyplot as pyplot

pyplot.figure(figsize=(10,5))

pyplot.rcParams.update({'font.size':18})

pyplot.xlabel('yhat-y')

pyplot.ylabel('Error')

pyplot.plot(yhat_minus_y, absolute_error)

pyplot.plot(yhat_minus_y, mean_squared_error)

pyplot.legend(['Absolute Error', 'Squared Error'])
```
