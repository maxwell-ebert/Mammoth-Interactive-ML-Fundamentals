#Exploring distributions!

1. Uniform Distributions

```python
import numpy

data = numpy.random.uniform(size = 1000)

import seaborn

graph = seaborn.displot(data)
```

2. Gaussian/Normal Distributions

```python
import numpy

data = numpy.random.normal(size = 1000)

import seaborn

graph = seaborn.displot(data)
```

3. Log-Normal Distribution

```python
import numpy

data = numpy.random.lognormal(size = 1000)

import seaborn

data_graph = seaborn.displot(data)
```

4. Exponential Distribution



```python
import numpy

data = numpy.random.exponential(size = 1000)

import seaborn

graph = seaborn.displot(data)
```

5. Laplace Distribution

```python
import numpy

data = numpy.random.laplace(size = 1000)

import seaborn

graph = seaborn.displot(data)
```

6. Binomial Distribution

```python
import numpy

number_of_trials = 5 

number_of_experiments = 1000

probability_of_heads = 0.5

count_of_heads = numpy.random.binomial(number_of_trials, 
                      probability_of_heads, 
                      number_of_experiments)

heads, event_count = numpy.unique(count_of_heads, return_counts = True)

probability_of_heads_in_experiments = event_count/number_of_experiments

print(probability_of_heads_in_experiments)
```

```python
import matplotlib.pyplot as pyplot

pyplot.figure(figsize = (15,10))

pyplot.bar(heads, probability_of_heads_in_experiments)

pyplot.ylabel('Event Probability')

pyplot.xlabel('Head Flips (Out of 5')
```
