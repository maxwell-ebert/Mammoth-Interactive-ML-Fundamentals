6. Binomial Distribution

```python
import numpy

number_of_trials = 60 

number_of_experiments = 100000

probability_of_heads = 0.5

count_of_heads = numpy.random.binomial(number_of_trials, 
                      probability_of_heads, 
                      number_of_experiments)

heads, event_count = numpy.unique(count_of_heads, return_counts = True)

probability_of_heads_in_experiments = event_count/number_of_experiments

print(probability_of_heads_in_experiments)
```

```python
import matplotlib.pyplot as pyplot

pyplot.figure(figsize = (15,10))

pyplot.rcParams.update({'font.size' : 18})

pyplot.bar(heads, probability_of_heads_in_experiments)

pyplot.ylabel('Event Probability')

pyplot.xlabel('Head Flips (Out of 5')
```

7. Multinomial Distribution

```python
import numpy

number_of_experiments = 1000

rolls = numpy.random.multinomial(number_of_experiments, [1/6.]*6)

event_probability = rolls/number_of_experiments

print(event_probability)


```

```python
import matplotlib.pyplot as pyplot

pyplot.figure(figsize=(10,5))

pyplot.rcParams.update({'font.size' : 18})

pyplot.xlabel('Die Roll')

pyplot.ylabel('Event Probability')

pyplot.bar(range(1,7), event_probability)
```

8. Poisson Distribution

```python
from numpy.ma.core import true_divide
import numpy

lam = 5

number_of_samples = 1000

samples = numpy.random.poisson(lam, number_of_samples)

x, x_count = numpy.unique(samples, return_counts = True)

probability = x_count / number_of_samples

print(probability)
```

```python
import matplotlib.pyplot as pyplot
pyplot.figure(figsize=(10,5))
pyplot.rcParams.update({'font.size': 18})
pyplot.bar(x,probability)

pyplot.xlabel('X')
pyplot.ylabel('Probability')
```
